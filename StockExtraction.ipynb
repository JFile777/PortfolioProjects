```python
import yfinance as yf
import pandas as pd
# import matplotlib inline
```

```python
# Using the Ticker Module we can create an object that allows us to access functions to extract data, 
# need to provide the ticker symbol for the stock
apple = yf.Ticker('AAPL')
```

```python
#Linux !wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json

import urllib.request

# URL of the file you want to download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json"

# Local filename where the file will be saved
filename = "apple.json"

# Download the file
urllib.request.urlretrieve(url, filename)

print("Download complete!")


```

```python
import json

#open the JSON file and laod the data into a var
with open("apple.json") as json_file:
    apple_info = json.load(json_file)
    #Print the type of data variable
    #print("Type:", type(apple info))
apple_info
```

```python
# We can get the country keyword using 'country'

apple_info['country']
```

```python
# We can use the history() method to get the share price of stock over certain interval,
# Using the period parameter we can set how far back from the present to get the data
# Options ofr period are 1day(1d),3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.
apple_share_price_data = apple.history(period='max')

# The format that the data is returned in is a pandas dataframe, with Date as the index, the stock Open,High,Low,Close,Volume, and Splits are given/day

apple_share_price_data.head()
# Wow looks so good and neat I like
```

```python
# We can reset the index of the df with the reset_index function

apple_share_price_data.reset_index(inplace=True)

apple_share_price_data.head()

# We also set the inplace param to True so the change takes place to the df itself
```

```python
# We can plot the Open price against the Date

apple_share_price_data.plot(x="Date", y="Open")
```

```python
# We can extract dividends as so:

apple.dividends

apple.dividends.plot(x='Date',y='Dividends')
```

```python
# Obtain stock data for AMD using the `history` function, set the `period` to max. 
# Find the `Volume` traded on the first day (first row).



amd = yf.Ticker("AMD")

#Get historical market data, setting period to max


amd_share_price_data = amd.history(period='max')

amd_share_price_data.head()
```

```python
# Find the `Volume` traded on the first day (first row)

# To store the assignment
first_day_volume = amd_share_price_data.iloc[0]['Volume']

# Could also do this, but better practice is to store the value in var

amd_share_price_data.iloc[0]['Volume']



first_day_volume
```
